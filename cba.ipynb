```python
import pandas as pd
import numpy as np
```

```python
def outsourcing(hashtable):
    hashTable = hashtable[hashtable.columns[0:2]] 
    hashDic = hashTable.set_index('Parameter')['hash value'].to_dict()
    
    #read in data based on year
    #year: 0 (2018); 1 (2019)
    year = hashDic.get('year')
    if year == 0:
        data = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='Volume 2018')
    elif year == 1:
        data = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='Volume 2019')
    
    #sourcing types: 0 (outsourcing); 1 (insourcing); 2 (hybrid)
    sourcing = hashDic.get('sourcing')
    
    
    #outsourcing
    #total outsourcing cost = cost provided by Abbott + shipping cost
    location = hashDic.get('location')
    
    #account type: 0 (US Commercial); 1 (US R&D); 2 (EU Commercial); 3 (EU R&D)
    accountType = hashDic.get('account type') 
 
    if accountType == 0:
        account = 'US Commercial'
    elif accountType == 1:
        account = 'US R&D'
    elif accountType == 2:
        account = 'EU Commercial'
    elif accountType == 3:
        account = 'EU R&D'
        
    prio = hashDic.get('priority') 
    if prio == 0:
        priority = 'Standard' 
    elif prio == 1:
        priority = 'Top Priority'
    elif prio == 2:
        priority = 'Rush'

    
    #cost provided by Abbott
    df = data.loc[(data['Location'] == location) & (data['Account Type'] == account)]
    
    if priority == 0 or 1 or 2:
        df = data.loc[(data['Location'] == location) & (data['Account Type'] == account) & (data['Test Priority'] == priority)]
        
    #toal volume
    df_testV = df.groupby(['Test Code'])['Cost Item List Price'].count().reset_index()
    df_testV.rename({'Test Code': 'Test Code', 'Cost Item List Price': 'Test Volume'}, axis=1, inplace=True)
    print(df_testV)
        
    #total cost
    df_testC = df.groupby(['Test Code'])['Cost Item List Price'].sum().reset_index()
    df_testC.rename({'Test Code': 'Test Code', 'Cost Item List Price': 'Test Cost'}, axis=1, inplace=True)
    print(df_testC)
    
    #shipping cost
    #shippingCost = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='shipping')
    #origin_location = hashDic.get('origin location') 
    #test_location = hashDic.get('test location') 
    #origin_test_location = (origin_location, test_location)
    
    #shippingDic = dict([((a,b), i) for a,b, i in zip(shippingCost['from'],shippingCost['to'], shippingCost['cost'],)])
    #shipping_cost = shippingDic.get(origin_test_location)
    
    #total_outsourcing_cost = cost + shipping_cost
    
    #print('Total Volume is: ' + str(volume))
    #print('Total Cost is: ' + str(total_outsourcing_cost))

```

```python
hashtable = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='hashtable')
outsourcing(hashtable)
```

```python
def cbaModel(hashtable):
    
    #hash table
    hashTable = hashtable[hashtable.columns[0:2]] 
    hashDic = hashTable.set_index('Parameter')['hash value'].to_dict()
    
    #read in data based on year
    #year: 0 (2018); 1 (2019)
    year = hashDic.get('year')
    if year == 0:
        data = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='Volume 2018')
    elif year == 1:
        data = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='Volume 2019')
    
    #read in insourcing data
    insourcing_data = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='insourcing cost')
    
    #sourcing types: 0 (outsourcing); 1 (insourcing); 2 (hybrid)
    sourcing = hashDic.get('sourcing')
    
    #outsourcing
    #total outsourcing cost = cost provided by Abbott + shipping cost
    if sourcing == 0:
        #cost provided by Abbott
        #toal volume
        df_testV = data.groupby(['Test Code'])['Cost Item List Price'].count().reset_index()
        df_testV.rename({'Test Code': 'Test', 'Cost Item List Price': 'Outsourcing Test Volume'}, axis=1, inplace=True)
        
        #total cost
        df_testC = data.groupby(['Test Code'])['Cost Item List Price'].sum().reset_index()
        df_testC.rename({'Test Code': 'Test', 'Cost Item List Price': 'Outsourcing Test Cost'}, axis=1, inplace=True)
        
        
    #shipping cost
    #shippingCost = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='shipping')
    #origin_location = hashDic.get('origin location') 
    #test_location = hashDic.get('test location') 
    #origin_test_location = (origin_location, test_location)
    
    #shippingDic = dict([((a,b), i) for a,b, i in zip(shippingCost['from'],shippingCost['to'], shippingCost['cost'],)])
    #shipping_cost = shippingDic.get(origin_test_location)
    
    #total_outsourcing_cost = cost + shipping_cost
    
    #print('Total Volume is: ' + str(volume))
    #print('Total Cost is: ' + str(total_outsourcing_cost))
        
    
    #insourcing
    if sourcing == 1:
        service_contract = hashDic.get('service contract')
        #exclude the contract fee
        if service_contract == 0:
            insource_cost = pd.DataFrame(insourcing_data, columns=['Test','total cost without contract'])
            insource_cost.rename({'Test Code': 'Test', 'total cost without contract': 'Insourcing Test Cost'}, axis=1, inplace=True)
            print(insource_cost)
        
        #include the contract fee
        if servuce_contract == 1:
            insource_cost = pd.DataFrame(insourcing_data, columns=['Test','Cost total'])
            insource_cost.rename({'Test Code': 'Test', 'Cost total': 'Insourcing Test Cost'}, axis=1, inplace=True)
            print(insource_cost)
            
    #hybrid
    if sourcing == 2:
        df_testC = data.groupby(['Test Code'])['Cost Item List Price'].sum().reset_index()
        df_testC.rename({'Test Code': 'Test', 'Cost Item List Price': 'Outsourcing Test Cost'}, axis=1, inplace=True)
        
        service_contract = hashDic.get('service contract')
        
        if service_contract == 0:
            insource_cost = pd.DataFrame(insourcing_data, columns=['Test','total cost without contract'])
            insource_cost.rename({'Test Code': 'Test', 'total cost without contract': 'Insourcing Test Cost'}, axis=1, inplace=True)
        
        #include the contract fee
        if service_contract == 1:
            insource_cost = pd.DataFrame(insourcing_data, columns=['Test','Cost total'])
            insource_cost.rename({'Test Code': 'Test', 'Cost total': 'Insourcing Test Cost'}, axis=1, inplace=True)
        
        
        ratio = hashDic.get('ratio')
        new_df = df_testC.merge(insource_cost,how='left')
        new_df['ratio'] = new_df['Outsourcing Test Cost'] / new_df['Insourcing Test Cost'] 
        
        insourcing_test = new_df.loc[new_df['ratio'] > ratio] 
        print('Test that should be insourced')
        print(insourcing_test)
        
        outsourcing_test = new_df.loc[new_df['ratio'] < ratio] 
        print('Test that should be outsourced')
        print(outsourcing_test)
        
        
```

```python
hashtable = pd.read_excel (r'~/Desktop/Volume data by region_DTS + Sophia copy.xlsx', sheet_name='hashtable')
cbaModel(hashtable)
```
